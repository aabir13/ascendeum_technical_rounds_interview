```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('analysis_id(in).csv')
df.head(5)
```

```python
df.columns
```

```python
df = df[['bidder', 'request_count', 'response_count', 'prebid_win_count', 'sum_bid',      'sum_time_to_respond', 'median', 'min_bid', 'max_bid', 'avg_bid', 'sum_2nd_highest_bid', 'sum_prebid_winning_bid', 'sum_winning_bid', 'sum_nobid', 'sum_timeout']]
```

```python
# Clean the dataframe to replace null values & get rid of duplicates
df.drop_duplicates(inplace=True)
cleaned_df = df.fillna(0)
```

```python
cleaned_df.isna().sum()
```

```python
cleaned_df
```

```python
cleaned_df['bidder'].value_counts()
```

```python
# Create the metrics table for win rate analysis
metrics = (cleaned_df.groupby('bidder')
    .agg(
        total_bids = ('request_count', 'sum'),
        prebid_wins = ('prebid_win_count', 'sum'),
        responded_bids = ('response_count', 'sum'),
        sum_bid = ('sum_bid', 'sum'),
        min_bid = ('min_bid', 'min'),
        max_bid = ('max_bid', 'max'),
        avg_bid = ('avg_bid', 'mean'),
        sum_prebid_winning_bid = ('sum_prebid_winning_bid', 'sum')
    )
    .reset_index()
)
```

```python
# Calculate win-rate & absolute win-rate
metrics['win_rate'] = metrics['prebid_wins'] / metrics['total_bids']
metrics['abs_win_rate'] = metrics['prebid_wins'] / metrics['responded_bids']
```

```python
# Rank Wins
rank_wins = metrics.sort_values(by='win_rate', ascending=False)
rank_wins
```
